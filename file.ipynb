```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset
data = pd.read_csv('https://lukeb.co/python_csv')
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_exploded = df.explode('job_skills')

df_exploded['job_skills'].value_counts()
#df_exploded.value_counts('job_skills')
#df_exploded
```

```python
df_exploded
```

```python
skills = df_exploded.groupby(['job_skills','job_title_short']).size()
df_skills = skills.reset_index(name= 'Skills_count')
count = df_skills.sort_values(by= 'Skills_count', ascending= False)
count
```

```python
skills = df_exploded[['job_skills','job_title_short']].value_counts()
df_skills = skills.reset_index(name= 'Skills_count')
count = df_skills.sort_values(by= 'Skills_count',ascending= False)
count
```

```python
job_titles = ['Data Scientist','Data Engineer', 'Data Analyst']
df_plot1 = (count[count['job_title_short']== title] for title in job_titles)
df_plot = pd.concat(df_plot1)

df_plot5 = (
    df_plot
    .groupby('job_title_short')
    .head(5)
)
df_plot5.plot(kind= "barh", x= 'job_title_short', y= 'Skills_count')
df_plot
```

```python

fig, ax = plt.subplots(3,1)
df_exploded1 = df.explode('job_skills')
skill1 = df_exploded1[['job_skills','job_title_short']].value_counts()
d_skill = skill1.reset_index(name= 'Counting')
count1 = d_skill.sort_values(by= 'Counting', ascending= False)

job_titles = ['Data Scientist','Data Engineer', 'Data Analyst',]
for i, job_title in enumerate(job_titles):
    df_plot = count1[count1['job_title_short'] == job_title].head(5)
    df_plot.plot(kind='barh', x='job_skills', y='Counting', ax=ax[i], title= job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].legend().set_visible(False)
    ax[i].set_xlim(0,120000)

fig.suptitle('Counts of Top Skills in Job Postings', fontsize=15)
fig.tight_layout()
plt.show()
```

```python
d1 = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
d1= d1.dropna(subset= ['salary_year_avg'])
d1 = d1.explode(['job_skills'])
d_group = d1.groupby('job_skills')['salary_year_avg'].agg(['count','median'])
top_pay = d_group.sort_values(by='median', ascending= False).head(10)
top_pay
```

```python
top_skills = d_group.sort_values(by='count', ascending= False).head(10).sort_values(by='median', ascending= False)
top_skills
```

```python
fig, ax = plt.subplots(2,1)
top_pay[::-1].plot(kind= 'barh', y='median', ax = ax[0], legend= False)
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}k'))

top_skills[::-1].plot(kind= 'barh', y='median', ax = ax[1], legend= False)
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}k'))

ax[1].set_xlim(ax[0].get_xlim())

ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysts')
ax[1].set_xlabel('Median Salary (USD)')

ax[0].set_ylabel('')
ax[1].set_ylabel('')

fig.tight_layout()
plt.show()
```
